# **Caminata Aleatoria**

Tenemos una caminata unidimensional que en cada paso se mueve una unidad  a la derecha con probabilidad $p$ o a la izquierda con probabilidad $q$, tal que:

$$ p + q = 1 $$



Vamos a realizar una simulación de la caminata aleatoria donde podamos observar cómo cambian las trayectorias al modificar el valor de $p$.

La caminata aleatoria puede verse como la suma acumulada de variables aleatorias independientes:

$$ X_n = \sum_{i=1}^n \xi_i $$

donde $ \xi_i $ representa un paso aleatorio con distribución:

$$ P(\xi_i = +1) = p, \quad P(\xi_i = -1) = q $$

```python
# Importamos las librerías necesarias

import random as rd
import matplotlib.pyplot as plt
```

Definimos una función con parámetros $n$ y $p$ que nos regresará la trayectoria o recorrido total realizado por la simulación.

Esta función nos permitirá elegir valores para $p$, al igual que un número de pasos $n$.



_Para la función:_

Como se cumple $$p+q=1$$ calculamos la probabilidad de ir a la izquierda: $$q=p-1$$

```python
def caminatas(n, p):
  q = p - 1
  paso = 0  # incializamos un contador de los pasos que se van a realizar
  trayectoria = [] # creamos una lista donde se guardará la trayectoria o recorrido total

  for i in range(n):
    num = rd.random() # generamos un número aleatorio

# si el número aleatorioes menor que la probabilidad p,
# entonces el contador aumenta una unidad hacia la derecha
# en caso contrario, aumenta una unidad hacia la iquierda
    if num < p:
      paso += 1
    else:
      paso -= 1

    trayectoria.append(paso) # guardamos cada paso en la lista para la trayectoria

  return trayectoria
```

```python
# Definimos los parámetros n y p
n = 100
p = 0.3

# Mandamos llamar la función e imprimimos la trayectoria que nos arroja
visual = caminatas(n, p)
print(visual)
```

```python
# Graficamos para ver la trayectoria de forma visual
plt.plot(visual, label=f'Trayectoria', color='purple')
plt.title(f'Caminata Aleatoria con p = {p}')
plt.xlabel('Número de pasos')
plt.legend()
plt.grid(True)
plt.show()
```

## ¿Qué preguntas surgen a partir de los resultados obtenidos?

Podemos hacernos preguntas como las siguientes:

**1. ¿Cómo influye el valor de $p$ en la dirección del movimiento?**

Como respuesta, podríamos decir que el valor de $p$ determina el sesgo del movimiento. Si $p = 0.5$, es simétrica. Si $p > 0.5$, hay una tendencia hacia la derecha, y si $p < 0.5$, hay una tendencia hacia la izquierda.

**2. ¿Por qué cada trayectoria es diferente si se usan los mismos parámetros?**

Cada trayectoria es distinta porque el proceso es aleatorio. Aunque pogamos los mismos parámetros de $p$ y $n$, se generan números aleatorios diferentes en cada paso, los cuales nos van a indicar la dirección de los mismos.

**3. ¿Podemos predecir la posición final después de muchos pasos?**

No, no podriamos predecir con exactitud la posición final de las trayectorias dado la variabilidad de los parámtros y la los números aleatorios generados.


## Conclusiones

A partir de las simulaciones realizadas de la caminata aleatoria se pueden observar distintos comportamientos según el valor de la probabilidad $p$. Podemos decir que:

- Cuando $p=0.5$, la caminata es simétrica y no presenta una tendencia hacia la izquierda o la derecha.

- Cuando $p > 0.5$, se observa una tendencia creciente, es decir, las trayectorias tienden a desplazarse a la derecha.

- Cuando $p < 0.5$, la mayoría de las trayectorias tienden a desplazarse hacia la izquierda, con una tendencia decreciente.

De igual forma, podemos decir que, al aumentar el número de pasos $n$, la trayectoria se aleja del punto de origen $0$.
